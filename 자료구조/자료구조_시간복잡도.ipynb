#최댓값을 찾는 arrangemax

algorithm arrangemax(A,n):

input: n개의 정수를 받는 배열 A

output: A의 수중에서 최댓값 리턴

currentmax = A[0]

for i = 1 to n-1 do(돌린다)

if currentmax < A[i]:

  currnetmax = A[i]

return currnetmax

A = [3,-1,9,2,12] N = 5일때 arrangemax는?

i = 3

currentmax = 3 --------------> 1

3 > -1 ----------------> 2

3< 9 ------------------> 3

currentmax = 9 ------------> 4

9>2 -----------------> 5

9< 12 -------------> 6

currentmax = 12 ----------------> 7


총 기본연산 7회 진행

하지만 무한히 많은 입력과 n이 존재가능하다.

#Big-0 표기법: 알고리즘의 수행 시간

1. 4n+1이 2n-1보다 2배 느리다.

+ 최고 차항이 동일하고 상수가 2배차이나므로.

2. f: y=((2)/(3)) x^(2)-((3)/(2)) x+1은 n<5/3이면 4n+1보다 빠르다.
3. 하지만 n>5/3이 되면 4n+1보다 느리다.

2n-1과 4n+1은 최고차항이 n이므로, 선형적으로 증가. 

하지만 y=((2)/(3)) x^(2)-((3)/(2)) x+1은 x제곱 만큼 증가

즉 BIG-O표기법이란 함수값을 결정하는 최고차항 만으로 간단하게 표기.

T(n) = 2n -1

T(n) = 4n+1

T(n) = ((2)/(3)) n^(2)-((3)/(2)) n+1

에서 

O(n)
O(n)
O(n^(2))

으로 표기를 바꾸어서 할 수 있다.
