#스택 계산기 코드

## 연산자, 피연산자, 토큰

2 + 3 * 5 
--> '2' '+' '3' '*' '5'

이때 2는 피연산자(operand), +, *는 연산자(operator)
그리고 이 두개를 토큰(token)이라 정의한다.

## 이항연산자, 단항연산자

이항 연산자는 2 + 3처럼 2와 3같은 항이 2개 필요한 연산자를 말한다.

단항 연산자는 +3같은 모습을 말한다.

## infix, postfix 수식

infix란 2+3*5처럼 연산자가 피연산자 사이에 있는 수식을 의미하고

postfix는 2 3 5 + *와 같이 연산자가 피연산자 사이에 없는 수식을 의미한다.


##postfix를 만드는 방법

1. infix 수식에서 괄호를 친다.

  (2 + (3 * 5))
2. 연산자의 괄호의 오른쪽으로 연산자를 이동시킨다.

  (2 (3 5) *) +
3. 괄호를 제거한다.

  2 3 5 * +

4. 완성


### 3 * (2 + 5) * 4를 postfix로 만들어보자.

((3* (2 + 5)) * 4)에서

((3 (2 5) + ) * 4) *

3 2 5 + * 4 *
